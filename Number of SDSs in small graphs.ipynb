```python
# This is a program that determines the number of split dominating sets (SDS) in graphs having at most 6 vertices
# It also prints out the graph with the maximum number of SDSs. 
```

```python
# for n=4
from itertools import chain, combinations
from sage.all import graphs

def get_stats(G, n):
    dom_count = 0
    sds_count = 0
    vertices = G.vertices()
    all_v_set = set(vertices)
    
    for r in range(1, n + 1):
        for S in combinations(vertices, r):
            S_set = set(S)
            nbrs = set()
            for v in S:
                nbrs.update(G.neighbors(v))
            
            if (S_set | nbrs) == all_v_set:
                dom_count += 1
                if len(all_v_set - S_set) >= 2:
                    H = G.copy()
                    H.delete_vertices(S)
                    if not H.is_connected():
                        sds_count += 1
                        
    return dom_count, sds_count

n = 4 
all_graphs = [G for G in graphs.nauty_geng(str(n)) if G.is_connected()]

# To store results for later comparison
results = []

for idx, G in enumerate(all_graphs):
    d_cnt, s_cnt = get_stats(G, n)
    results.append({'graph': G, 'sds': s_cnt, 'dom': d_cnt, 'id': idx})
    
    print(f"Graph ID: {idx} | SDSs: {s_cnt}")
    G.show(figsize=2)
    print("-" * 30)

# --- FIND AND PRINT THE MAXIMAL GRAPHS ---

# Find the maximum number SDSs from our results
max_sds_value = max(r['sds'] for r in results)
extremal_graphs = [r for r in results if r['sds'] == max_sds_value]

print("\n" + "="*40)
print(f"When the number of vertices in G is {n},the maximum number of SDSs is {max_sds_value}.")
print(f"And the graph attaining this maximum value is")
#print(f"Number of graphs attaining this maximum: {len(extremal_graphs)}.")
#print("="*40 + "\n")

for item in extremal_graphs:
    #print(f"Extremal Graph with {item['sds']} SDSs:")
    # figsize=5 makes the final result much larger
    item['graph'].show(figsize=4, title=f"")
    #item['graph'].show(figsize=5, title=f"The extremal graph is:")
```
